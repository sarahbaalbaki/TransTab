## Join the datasets together 

```python
# first, add an ID for the first dataframe 
patient_ids = list(range(0, len(df3)))

# Add patient IDs to the DataFrame
df3.insert(0, 'Patient_ID', patient_ids)
df3.head()
```

```python
df3_new= df3.rename(columns={"gender": "Gender", "age": "AGE", "HbA1c_level": "HbA1c", "diabetes": "CLASS", "bmi": "BMI", "Patient_ID": "ID"})
df3_new
```

```python
df4["ID"] = range(df3_new['ID'].max() + 1, df3_new['ID'].max() + 1 + len(df4))
df4["ID"]
```

```python
df_combined = pd.concat([df3_new, df4], ignore_index=True)
df_combined.info()
```

```python
print(df_combined['CLASS'].unique())
for col in df_combined.columns:
    # Check if the column is "CLASS"
    if col == "CLASS":
        # Replace "N" with 0 and "Y" with 1
        df_combined[col].replace({"N": 0, "Y": 1, "P":1}, inplace=True)
print(df_combined['CLASS'].unique())
df_combined.head()
```

```python
df_combined['Gender'] = df_combined['Gender'].str.replace('Female', 'F').str.replace('Male', 'M')
df_combined.head()
```

```python
filtered_df = df_combined[df_combined['AGE'] < 5]
filtered_df
```

### Impute Numerical Columns based on mean value 

```python
# numeric_columns_to_fill = ['No_Pation', 'Cr', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL']
numeric_columns_to_fill=[]

for column in merged_df.columns:
    column_datatype = merged_df[column].dtype
    if column_datatype== "float64" or column_datatype== "int64":
        numeric_columns_to_fill.append(column)

for col in numeric_columns_to_fill:
    mean_value= merged_df[col].mean()
    merged_df[col].fillna(mean_value, inplace=True)

merged_df.to_csv('filled_mean.csv', index=False)
merged_df.info()
```

### Impute Categorical Columns based on presence or absence of variable

```python
print(merged_df['smoking_history'].unique())
merged_df['smoking_history'].fillna('No Info', inplace=True)
print(merged_df['smoking_history'].unique())
```

```python
print(merged_df['Gender'].unique())
nan_count = merged_df['Gender'].isna().sum()
print(nan_count)
print(len(merged_df))
print((nan_count/len(merged_df))*100)
```

```python
for col in merged_df.columns:
    # Check if the column is "CLASS"
    if col == "Gender":
        # Replace "N" with 0 and "Y" with 1
        merged_df[col].replace({0: 'F', 1: 'M'}, inplace=True)
```
